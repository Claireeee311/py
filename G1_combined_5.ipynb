### Question 5.21: Olympics Medals

In this question, consider the number of gold/silver/bronze medals won by the top five countries. <br>

In a single figure, plot a horizontal bar chart showing the number of gold, silver, and bronze medals won by each of the five countries, using a different color to represent each of the three different types of medals. Try to arrange the rows according to the number of medals won by each country (ie. top row corresponds to the country with the highest number of medals).

*P.S. This is not the real Olympics results.*

```python
medals = {
    'USA': [25, 31, 23], 'CHN': [32, 22, 16], 'AUS': [14, 21, 18], 'GBR': [15,18,15], 'JPN': [21, 7, 12]
}
```

```python
medal_df = pd.DataFrame.from_dict(
    medals, orient='index', columns = ['Gold' , 'Silver', 'Bronze']
)
```

```python
medal_df
```

```python
medal_df['Total'] = medal_df.apply(lambda x: x.sum(), axis=1)
medal_df.sort_values(by=['Total'], ascending=True, inplace=True)
medal_df
```

```python
fig = plt.figure()
plt.barh(medal_df.index, medal_df['Gold'], color='gold')
plt.barh(medal_df.index, medal_df['Silver'], left=medal_df['Gold'], color='silver')
plt.barh(medal_df.index, medal_df['Bronze'], left=medal_df['Gold']+medal_df['Silver'], color='Brown')
plt.legend(['Gold', 'Silver', 'Bronze'])
plt.show()
```

```python
fig = plt.figure()
plt.bar(medal_df.index, medal_df['Gold'], color='gold')
plt.bar(medal_df.index, medal_df['Silver'], bottom=medal_df['Gold'], color='silver')
plt.bar(medal_df.index, medal_df['Bronze'], bottom=medal_df['Gold']+medal_df['Silver'], color='Brown')
plt.legend(['Gold', 'Silver', 'Bronze'])
plt.show()
```

Create a pie chart to see the proportion of gold, silver and bronze medals won by each of the five countries.

```python
keys = list(medals.keys())
gold = [x[0] for x in medals.values()]
silver = [x[1] for x in medals.values()]
bronze = [x[2] for x in medals.values()]
```

```python
print(gold)
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,4))
axes[0].pie(gold)
axes[0].set(title = 'Gold')
axes[1].pie(silver)
axes[1].set(title = 'Silver')
axes[2].pie(bronze)
axes[2].set(title = 'Bronze')
axes[2].legend(keys)
# plt.tight_layout()
# plt.figlegend(keys, loc='upper left', fontsize='small')
plt.show()
```

```python
maxx = pop_by_region['Year2020'].max()
minn = pop_by_region['Year2020'].min()

print(maxx, minn)
```
