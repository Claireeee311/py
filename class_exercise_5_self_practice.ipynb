### Contour

```python
def f(x, y):
    return np.sin(np.pi*x)*np.sin(np.pi*y)
```

```python
x = np.linspace(0, 2, 500)
y = np.linspace(0, 2, 500)
X, Y = np.meshgrid(x, y) #create positions
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure() 
axes = fig.gca(projection ='3d') 
axes.plot_surface(X, Y, Z) 
axes.contour(X, Y, Z, offset = -1)
plt.show()
```

### Cluster

```python
# from sklearn.datasets.samples_generator import make_blobs
# The sklearn.datasets.samples_generator module is removed in version 0.24
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.50, random_state=0)
#Generate isotropic Gaussian blobs for clustering
#X: Generated positions
#y: The integer labels for cluster membership of each sample (position)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

### Cluster - KMeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red');
```

### Data Visualisation

```python
import pandas as pd
data = pd.read_csv('crime_data.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['Area ID'] = data['Area ID'].astype(str)
data['Crime Code'] = data['Crime Code'].astype(str)
```

```python
data.isnull().sum()
```

```python
crime_by_area = pd.DataFrame({'count': data.groupby('Area Name').size()}).reset_index()
crime_by_area = crime_by_area.sort_values('count')
plt.bar(crime_by_area['Area Name'], crime_by_area['count'])
plt.xticks(rotation=90)
plt.title('Crime count by Area')
plt.show()
```

```python
data = data[data['Area Name'] == '77th Street']
```

```python
crimes = pd.DataFrame({'count': data.groupby(['Crime Code'])['Crime Code'].count()}).reset_index()
crimes = crimes.sort_values('count')
crimes = crimes[crimes['count'] > 1000]
```

```python
plt.bar(crimes['Crime Code'], crimes['count'], color = 'salmon')
plt.xticks(rotation=90)
plt.title('Crime count by Type on 77th Street')
plt.show()
```

```python
top_crimes = data[data['Crime Code'].isin(['624', '510', '230'])]
```

```python
top_crimes['Hour Occurred'] = top_crimes.apply(lambda x: x['Time Occurred'][:-2] if len(x['Time Occurred']) > 2 else 0, axis = 1)
top_crimes['Hour Occurred'] = top_crimes['Hour Occurred'].astype(int)
```

```python
count_of_crimes = pd.DataFrame({'count': top_crimes.groupby(['Hour Occurred','Crime Code'])['Crime Code'].count()}).reset_index()
```

```python
count_of_crimes.head()
```

```python
crime_by_code_hour = count_of_crimes.pivot(index='Hour Occurred', columns='Crime Code', values='count')
```

```python
crime_by_code_hour
```

```python
for col in crime_by_code_hour.columns:
    plt.plot(crime_by_code_hour.index, crime_by_code_hour[col], label = col)
plt.legend()
plt.title('Crime occurred')
plt.show()
```

```python
cols = crime_by_code_hour.columns
for col in range(len(cols)):
    if col == 0:
        plt.bar(crime_by_code_hour.index, crime_by_code_hour[cols[col]], 
                label = cols[col])
        bot = crime_by_code_hour[cols[col]]
    else:
        plt.bar(crime_by_code_hour.index, crime_by_code_hour[cols[col]], 
                label = cols[col], bottom = bot)
        bot += crime_by_code_hour[cols[col]]
        
plt.title('Crime occurred')        
plt.legend()
plt.show()
```

```python
size = ['x', 'y', 'z']
labels = ['length', 'width', 'height']
for i in range(3):
    sns.histplot(x = size[i], 
                 data = diamonds, 
                 kde = True, 
                 color = sns.color_palette('Set1')[i], 
                 label = labels[i], bins = 10)
plt.legend()
plt.xlabel('mm')
plt.show()
```
